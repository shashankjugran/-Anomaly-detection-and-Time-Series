```python
'''
Question 4: Define Stationary in time series. How can you test and transform a non-stationary series into a stationary one?
  Answer-
  Definition of Stationarity

A time series is said to be stationary if its statistical properties remain constant over time. These properties include:

Mean (average) → constant over time
Variance → constant over time
Autocorrelation → depends only on lag, not time
Why Stationarity is Important

Most time series models like AR, MA, and ARIMA assume the data is stationary.
If the data is not stationary:

Model predictions become unreliable
Patterns like trend and seasonality distort results
Types of Stationarity
1. Strict Stationarity

All statistical properties remain constant (rare in real data).

2. Weak Stationarity (Most Important)

Only mean, variance, and covariance are constant.

How to Test Stationarity
1. Visual Inspection

Plot the data and check:

Trend (increasing/decreasing)
Seasonality (repeating patterns)
Variance changes
2. Augmented Dickey-Fuller (ADF) Test

It is a statistical test used to check stationarity.

Null Hypothesis (H₀): Series is non-stationary
Alternative Hypothesis (H₁): Series is stationary
Decision Rule:
p-value < 0.05 → Stationary
p-value ≥ 0.05 → Non-stationary
'''
import pandas as pd
import numpy as np
from statsmodels.tsa.stattools import adfuller

# Load dataset from online source
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/airline-passengers.csv"
data = pd.read_csv(url)

# Convert column
data['Month'] = pd.to_datetime(data['Month'])
data.set_index('Month', inplace=True)

# Define series
series = data['Passengers']

# ADF Test
result = adfuller(series)

print("ADF Statistic:", result[0])
print("p-value:", result[1])

if result[1] < 0.05:
    print("Series is Stationary")
else:
    print("Series is Non-Stationary")

#How to Convert Non-Stationary to Stationary

#If the series is non-stationary, apply the following transformations:

#1. Differencing (Most Common Method)

#Removes trend by subtracting previous value.
series_diff = series.diff().dropna()

#2. Log Transformation
#Reduces variance and stabilizes fluctuations.
import numpy as np
series_log = np.log(series)

#3. Moving Average
#Smooths the series
series_ma = series.rolling(window=12).mean()

#4. Seasonal Differencing
#Removes seasonal patterns
series_seasonal_diff = series.diff(12).dropna()

#Re-testing After Transformation
result = adfuller(series_diff)
print("New p-value:", result[1])

#Example Explanation
#The AirPassengers dataset shows:
#Increasing trend
#Seasonal pattern
#After applying differencing, the trend is removed and the series becomes stationary
```

```python
'''
Question 6: Load a time series dataset (e.g., AirPassengers), plot the original series, and decompose it into trend, seasonality, and residual components
'''
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

# Load dataset from online source
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/airline-passengers.csv"
data = pd.read_csv(url)

# Convert to datetime
data['Month'] = pd.to_datetime(data['Month'])

# Set index
data.set_index('Month', inplace=True)

# Rename column for simplicity
data.rename(columns={'Passengers': 'passengers'}, inplace=True)

# Plot original time series
plt.figure(figsize=(10,5))
plt.plot(data['passengers'])
plt.title("Original Time Series")
plt.xlabel("Year")
plt.ylabel("Passengers")
plt.show()

# Decompose the time series
decomposition = seasonal_decompose(data['passengers'], model='multiplicative')

# Plot decomposition
decomposition.plot()
plt.show()
```

```python
'''
Question 7: Apply Isolation Forest on a numerical dataset (e.g., NYC Taxi Fare) to
detect anomalies. Visualize the anomalies on a 2D scatter plot.
'''
from sklearn.ensemble import IsolationForest
import pandas as pd
import matplotlib.pyplot as plt

# Sample dataset
data = pd.DataFrame({
    'fare': [10, 12, 11, 300, 13, 14, 500],
    'distance': [1, 1.2, 1.1, 10, 1.3, 1.4, 15]
})

model = IsolationForest(contamination=0.2)
data['anomaly'] = model.fit_predict(data)

# Plot
plt.scatter(data['fare'], data['distance'], c=data['anomaly'])
plt.xlabel("Fare")
plt.ylabel("Distance")
plt.title("Isolation Forest Anomalies")
plt.show()
```

```python
'''
Question 8: Train a SARIMA model on the monthly airline passengers dataset.
Forecast the next 12 months and visualize the results.
'''
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX

# Load dataset from external source
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/airline-passengers.csv"
data = pd.read_csv(url)

# Convert to datetime
data['Month'] = pd.to_datetime(data['Month'])

# Set index
data.set_index('Month', inplace=True)

# Rename column
data.rename(columns={'Passengers': 'passengers'}, inplace=True)

# Train SARIMA model
model = SARIMAX(data['passengers'],
                order=(1,1,1),
                seasonal_order=(1,1,1,12))

result = model.fit()

# Forecast next 12 months
forecast = result.forecast(steps=12)

# Plot results
plt.figure(figsize=(10,5))
plt.plot(data['passengers'], label='Original')
plt.plot(forecast, label='Forecast', linestyle='--')
plt.title("SARIMA Forecast (Next 12 Months)")
plt.xlabel("Year")
plt.ylabel("Passengers")
plt.legend()
plt.show()
```

```python
'''
Question 9: Apply Local Outlier Factor (LOF) on any numerical dataset to detect
anomalies and visualize them using matplotlib.
'''
from sklearn.neighbors import LocalOutlierFactor
import matplotlib.pyplot as plt
import numpy as np

# Data
X = np.array([[1,1],[1.1,1.2],[0.9,1.1],[10,10],[10.5,10.2]])

lof = LocalOutlierFactor(n_neighbors=2)
y_pred = lof.fit_predict(X)

plt.scatter(X[:,0], X[:,1], c=y_pred)
plt.title("LOF Anomaly Detection")
plt.show()
```

```python
'''
Question 10: You are working as a data scientist for a power grid monitoring company.
Your goal is to forecast energy demand and also detect abnormal spikes or drops in
real-time consumption data collected every 15 minutes. The dataset includes features
like timestamp, region, weather conditions, and energy usage.
Explain your real-time data science workflow:
● How would you detect anomalies in this streaming data (Isolation Forest / LOF /
DBSCAN)?
● Which time series model would you use for short-term forecasting (ARIMA /
SARIMA / SARIMAX)?
● How would you validate and monitor the performance over time?
● How would this solution help business decisions or operations?

    answer-
    1. Anomaly Detection in Streaming Data
Use Isolation Forest for real-time speed
Use LOF for local anomalies
Use DBSCAN for clustering anomalies

Preferred: Isolation Forest (fast & scalable)

2. Forecasting Model
Use SARIMAX
Handles seasonality
Includes weather & region (external variables)
3. Validation & Monitoring
Metrics:
MAE (Mean Absolute Error)
RMSE
Techniques:
Rolling window validation
Real-time dashboards
Drift detection
4. Business Impact
Prevent power outages
Optimize energy distribution
Reduce operational cost
Improve demand planning
'''
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest
from statsmodels.tsa.statespace.sarimax import SARIMAX

# Create sample dataset (since real dataset not provided)
date_range = pd.date_range(start='2023-01-01', periods=200, freq='H')

data = pd.DataFrame({
    'timestamp': date_range,
    'energy': np.random.normal(100, 10, 200)
})

# Add anomalies manually
data.loc[::20, 'energy'] = data['energy'] * 3

# Set index
data.set_index('timestamp', inplace=True)

# -----------------------
# Anomaly Detection
# -----------------------
model = IsolationForest(contamination=0.05)
data['anomaly'] = model.fit_predict(data[['energy']])

# Plot anomalies
plt.figure(figsize=(10,5))
plt.plot(data['energy'], label='Energy')
plt.scatter(data.index, data['energy'],
            c=data['anomaly'], cmap='coolwarm', label='Anomaly')
plt.legend()
plt.title("Anomaly Detection")
plt.show()

# -----------------------
# Forecasting (SARIMA)
# -----------------------
model = SARIMAX(data['energy'], order=(1,1,1), seasonal_order=(1,1,1,24))
result = model.fit()

forecast = result.forecast(steps=24)

plt.figure(figsize=(10,5))
plt.plot(data['energy'], label='Original')
plt.plot(forecast, label='Forecast', linestyle='--')
plt.legend()
plt.title("Energy Forecast")
plt.show()
```
